Exercícios 1 a 4
1) Peça um número e verifique se é maior que 10.
2) Peça um número e verifique se é negativo.
3) Peça um número e verifique se é igual a 100.
4) Peça um número e informe se ele é par ou ímpar.


```python
numero = float(input("Digite um número: "))

if numero > 10:
    print("O número é maior que 10.")
else:
    print("O número não é maior que 10.")
```

```python
numero = float(input("Digite um número: "))

if numero < 0:
    print("O número é negativo.")
else:
    print("O número não é negativo.")
```

```python
numero = int(input("Digite um número: "))

if numero == 100:
    print("O número é igual a 100.")
else:
    print("O número não é igual a 100.")
```

```python
numero = int(input("Digite um número inteiro: "))

if numero % 2 == 0:
    print("O número é par.")
else:
    print("O número é ímpar.")
```

Exercícios 5 a 7
5) Verifique se um número é múltiplo de 3.
6) Verifique se um número é múltiplo de 5.
7) Verifique se um número é divisível por 2 e por 3.

```python
numero = int(input("Digite um número: "))

if numero % 3 == 0:
    print("O número é múltiplo de 3.")
else:
    print("O número não é múltiplo de 3.")
```

```python
numero = int(input("Digite um número: "))

if numero % 5 == 0:
    print("O número é múltiplo de 5.")
else:
    print("O número não é múltiplo de 5.")
```

```python
numero = int(input("Digite um número: "))

if numero % 2 == 0 and numero % 3 == 0:
    print("O número é divisível por 2 e por 3.")
else:
    print("O número não é divisível por 2 e por 3.")
```

Exercícios 8 a 10
8) Verifique se um número está entre 10 e 50.
9) Verifique se um número é par e múltiplo de 5.
10) Peça um número e informe se ele é:
 múltiplo de 2
 múltiplo de 3
 múltiplo de 5
 nenhum dos anteriores

```python
numero = float(input("Digite um número: "))

if 10 <= numero <= 50:
    print("O número está entre 10 e 50.")
else:
    print("O número não está entre 10 e 50.")
```

```python
numero = int(input("Digite um número: "))

if numero % 2 == 0 and numero % 5 == 0:
    print("O número é par e múltiplo de 5.")
else:
    print("O número não atende às duas condições.")
```

```python
numero = int(input("Digite um número: "))

if numero % 2 == 0:
    print("Múltiplo de 2")

if numero % 3 == 0:
    print("Múltiplo de 3")

if numero % 5 == 0:
    print("Múltiplo de 5")

if numero % 2 != 0 and numero % 3 != 0 and numero % 5 != 0:
    print("Nenhum dos anteriores")
```
